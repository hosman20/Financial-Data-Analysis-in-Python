```python
valid_close_df = merged_df.dropna(subset=['Close_A', 'Close_B', 'Close_C'], how='any')
```

```python
valid_close_complete = valid_close_df.fillna(method='backfill')
```

```python
valid_close_complete.isnull().any(axis=1).sum()
```

```python
pd.set_option('display.max_rows', None)
```

```python
valid_close_complete.describe()
```

```python
pd.set_option('display.max_rows', 2)
```

```python
prices_only = valid_close_complete.drop(['Volume_A', 'Volume_B', 'Volume_C'], axis=1)
```

```python
prices_only
```

```python
valid_close_complete['Open_A'].plot(figsize=(12,6), linestyle='--', color='black', legend='Open_A')
valid_close_complete['Close_A'].plot(figsize=(12,6), linestyle='-', color='grey', legend='Close_A')
valid_close_complete['Low_A'].plot(figsize=(12,6), linestyle=':', color='black', legend='Low_A')
valid_close_complete['High_A'].plot(figsize=(12,6), linestyle='-.', color='grey', legend='High_A')
```

```python
valid_close_complete['Open_C'].plot(figsize=(12,6), linestyle='--', color='yellow', legend='Open_C')
valid_close_complete['Close_C'].plot(figsize=(12,6), linestyle='-', color='blue', legend='Close_C')
valid_close_complete['Low_C'].plot(figsize=(12,6), linestyle=':', color='green', legend='Low_C')
valid_close_complete['High_C'].plot(figsize=(12,6), linestyle='-.', color='red', legend='High_C')
```

```python
no_outlier_prices = prices_only[(np.abs(stats.zscore(prices_only)) < 6).all(axis=1)]
```

```python
no_outlier_prices['Open_C'].plot(figsize=(12,6), linestyle='--', color='yellow', legend='Open_C')
no_outlier_prices['Close_C'].plot(figsize=(12,6), linestyle='--', color='blue', legend='Close_C')
no_outlier_prices['Low_C'].plot(figsize=(12,6), linestyle='--', color='green', legend='Low_C')
no_outlier_prices['High_C'].plot(figsize=(12,6), linestyle='--', color='red', legend='High_C')
```

```python
pd.set_option('display.max_rows', None)
```

```python
no_outlier_prices[['Open_C', 'Close_C', 'Low_C', 'High_C']].describe()
```

```python
close_prices = no_outlier_prices[['Close_A', 'Close_B', 'Close_C']]
```

```python
delta_close_prices = (close_prices.shift(-1) - close_prices).fillna(0)
delta_close_prices.columns = ['Delta_Close_A', 'Delta_Close_B', 'Delta_Close_C']
delta_close_prices
```

```python
pd.set_option('display.max_rows', None)
delta_close_prices.describe()
```

```python
delta_close_prices['Delta_Close_A'].plot(kind='hist', bins=100, figsize=(12,6), color='blue', grid=True)
```

```python
delta_close_prices['Delta_Close_B'].plot(kind='box', figsize=(12,6), color='red', grid='True')
```

```python
pd.plotting.scatter_matrix(delta_close_prices, figsize=(10,10), color='blue', alpha=0.75, diagonal='kde', grid=True)
```
